##Ejercicio 1 - Funciones lambdas


Definir las siguientes funciones usando lambdas.

```python
#a. Crear una lambda que reciba un número y devuelva True si es múltiplo de 3, False en caso contrario.
multiplo_de_3 = lambda num: num % 3 == 0
print(f"6 es multiplo de 3? {multiplo_de_3(6)}")
print(f"7 es multiplo de 3? {multiplo_de_3(7)}")

#b. Crear una lambda que reciba un número y devuelva su cubo.
cubo = lambda num: num ** 3
print(f"El cubo de 3 es {cubo(3)}")
print(f"El cubo de 5 es {cubo(5)}")

#c. Crear una lambda que reciba dos números y devuelva su producto.
producto = lambda num1, num2: num1 * num2
print(f"El producto de 4 y 5 es {producto(4, 5)}")
print(f"El producto de 10 y 2 es {producto(10, 2)}")

#d. Crear una lambda que reciba dos números y devuelva el mayor.
mayor = lambda num1, num2: num1 if num1 > num2 else num2
print(f"El mayor entre 8 y 12 es {mayor(8, 12)}")
print(f"El mayor entre 20 y 7 es {mayor(20, 7)}")

#e. Crear una lambda que reciba una palabra y devuelva True si empieza con la letra “A" (o “a”).
empieza_con_a = lambda palabra: palabra.lower().startswith('a')
print(f"'Avion' empieza con 'A'? {empieza_con_a('Avion')}")
print(f"'casa' empieza con 'A'? {empieza_con_a('casa')}")
print(f"'arbol' empieza con 'A'? {empieza_con_a('arbol')}")

#f. Crear una lambda que reciba una temperatura en Celsius y la convierta a Fahrenheit.
celsius_a_fahrenheit = lambda celsius: (celsius * 9/5) + 32
print(f"25 grados celsius en Fahrenheit es {celsius_a_fahrenheit(25)}F")
print(f"0 grados celsius en Fahrenheit es {celsius_a_fahrenheit(0)}F")
```

Realizar el siguiente ejercicio usando lambdas

- una que duplique un número
- una que le sume 10
- una que calcule su cuadrado

```python
operaciones = [
    lambda x: x * 2,
    lambda x: x + 10,
    lambda x: x ** 2
]

num = int(input("Ingrese un numero: "))

print(operaciones[0](num))
print(operaciones[1](num))
print(operaciones[2](num))
```
