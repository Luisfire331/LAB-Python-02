# Ejercicio 6 - Minimax
Dado el juego conocido como NIM, implementar una solución a este usando el algoritmo minimax visto en clase

En este juego hay una pila de piedras (por ejemplo 10).

En cada turno un jugador puede quitar 1, 2 o 3 piedras.

Pierde el jugador que no puede mover (porque ya no quedan piedras).

MAX = Computadora

MIN = Humano

Dado lo anterior y el algoritmo visto en clase, modelar el juego y simularlo donde MAX sera la cpu y MIN el humado el cual interactua mediante el input, el primer turno sera el de MIN

























```python
import math

# MAX = 1 (CPU)
# MIN = -1 (Humano)

def utilidad(piedras, turno):
    return -math.inf if turno == 1 else math.inf


def minimax(piedras, turno):
    """
    Devuelve (valor, movimiento)
    movimiento siempre será 1, 2 o 3 (nunca None)
    """
    if piedras == 0:
        return utilidad(piedras, turno), None

    movimientos = [m for m in [1,2,3] if piedras - m >= 0]

    if not movimientos:
        return utilidad(piedras, turno), None

    # MAX
    if turno == 1:
        mejor_valor = -math.inf
        mejor_mov = movimientos[0]

        for m in movimientos:
            valor, _ = minimax(piedras - m, -1)
            if valor > mejor_valor:
                mejor_valor = valor
                mejor_mov = m

        return mejor_valor, mejor_mov

    # MIN
    else:
        mejor_valor = math.inf
        mejor_mov = movimientos[0]

        for m in movimientos:
            valor, _ = minimax(piedras - m, 1)
            if valor < mejor_valor:
                mejor_valor = valor
                mejor_mov = m

        return mejor_valor, mejor_mov


def jugar_nim(piedras_iniciales=10):
    piedras = piedras_iniciales
    turno = -1  # humano inicia

    print("=== JUEGO NIM ===")
    print("Quedan", piedras, "piedras.\n")

    while True:

        # Fin del juego
        if piedras == 0:
            if turno == -1:
                print("No puedes mover. ¡GANÓ LA CPU!")
            else:
                print("La CPU no puede mover. ¡GANASTE!")
            break

        if turno == -1:
            print("\nTu turno. Piedras:", piedras)
            mov = int(input("¿Cuántas quitas? (1, 2 o 3): "))

            while mov not in [1,2,3] or mov > piedras:
                mov = int(input("Inválido. Intenta de nuevo: "))

            piedras -= mov
            turno = 1

        else:
            print("\nTurno CPU. Pensando...")

            _, mov = minimax(piedras, 1)

            print("La CPU quita", mov, "piedras.")
            piedras -= mov
            turno = -1

        print("Piedras restantes:", piedras)


jugar_nim(10)
```
